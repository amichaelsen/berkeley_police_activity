### Potential Police Stops and End of Shift Analysis

As mentioned in an episode of Brooklyn 99, cops may initiate frivolous arrests close to the end of their shifts in order to work past their shift and receive overtime. These overtime incentives are explicitly called out by the [Brunetti Law Firm](http://brunettilaw.com/police-overtime-incentive) in New York. More locally in the Bay Area, an [article](https://oaklandnorth.net/2021/04/28/oakland-police-overtime-payments/) in 2021 documented the exceptional amount of overtime pay received by Oakland PD Officers, although no concrete causes of this pay are discerned, but several are discussed. 

The goal of this notebook is to examine the Berkeley PD stop data to see if there are any times of day where a spike in minor crime calls are made, potentially indicating a similar end-of-shift arrest spree. A brief glimpse into the publicly reported Oakland PD data demonstrates why such an analysis cannot be applied there. 


#### The Data:
The data comes from [Berkeley Police Department Stop data](https://data.cityofberkeley.info/Public-Safety/Berkeley-PD-Stop-Data-Jan-26-2015-to-Sep-30-2020-/4tbf-3yt8/data) on the City of Berkeley [Open Data and Public Records](https://www.cityofberkeley.info/DataPortalTabs/) website.

The City of Berkeley official description can be found [here](https://data.cityofberkeley.info/api/views/4tbf-3yt8/files/2ed1e25d-effc-4e3f-b74c-21599b22b85f?download=true&filename=Berkeley%20Stop%20Data.pdf)

Description PDF at `Berkeley_Stop_Data.pdf`

```python
# import packages
import pandas as pd 
import matplotlib.pyplot as plt
import re
```

```python
# load in the data from Berkeley-PD-calls-for-service.csv
stops = pd.read_csv("Berkeley_PD_-_Stop_Data.csv")
```

```python
stops.head()
```

```python
stops[['Reason','CallType','City']].groupby(['Reason','CallType']).count()
```

```python
#investigate entries with no 'reason'
stops[stops['Reason']=='0']
#lots of missing data for these entries, but does have time data. 
```

```python
stops['CallType'].value_counts()
#T = traffic
# 1194 - pedestrian
# 1196 - suspicious vehicle stop
# 1194B - bicycle stop 
```

```python
# extract time of day information from 'CreateDatetime' column
sample = stops.iloc[0,0]
print(sample)
pattern = r"(\d*)/(\d*)/(\d*) (\d*):(\d*):(\d*) (A|P)M"
groups = re.search(pattern, sample)
for i in range(8):
    print(i,": ",groups.group(i))
```

```python
stops['datetime'] = pd.to_datetime(stops['CreateDatetime'])
```

```python
stops['hour']   = stops['datetime'].dt.hour
stops['minute'] = stops['datetime'].dt.minute
stops['year']   = stops['datetime'].dt.year
```

```python
stops.head()
```

```python
stops['timeInMinutes'] = stops['hour']*60 + stops['minute']
```

```python
stops.head()
```

```python
recent_stops = stops[stops['year']>=2018]
recent_stops.head()
```

```python
def plot_hist_of_stops_in_day(df, col_name="timeInMinutes", bins_per_hour=4):
    """
    Takes a dataframe with column listing the time of day in minutes (e.g. 2:33pm = 14*24+33) and plots 
    a density histogram of the entries by time of day with bins_per_hour bins for each hour of the day. 
    """
    # Make a histogram of time of stops by where in the day they occurred
    ax = df[[col_name]].hist(figsize=(18,7),bins=24*bins_per_hour, density=True, range=(0,24*60))[0][0]
    ax.set_xticks(list(range(0,24*60,60)));
    hour_labels = ["12:00AM", "1:00AM", "2:00AM", "3:00AM", "4:00AM", "5:00AM", 
                    "6:00AM", "7:00AM", "8:00AM", "9:00AM", "10:00AM", "11:00AM", 
                    "12:00PM", "1:00PM", "2:00PM", "3:00PM", "4:00PM", "5:00PM", 
                    "6:00PM", "7:00PM", "8:00PM", "9:00PM", "10:00PM", "11:00PM"]
    ax.set_xticklabels(hour_labels);
    plt.xticks(rotation = 45);
    ax.set_title("Density of Time of Day Stop Occurred")
    ax.set_xlim(0,24*60);
    
def density_by_minute_in_hour(df, col_name="minute", bins_per_hour=4):
    """
    Takes a dataframe with column listing the in minutes past the hour (e.g. 2:33pm = 33 minutes)
    and plots a density histogram of the entries by by minute with bins_per_hour bins for the hour.
    """
    ax = df[[col_name]].hist(figsize=(18,7),bins=bins_per_hour, density=True, range=(0,60))[0][0]
    ax.set_xticks(list(range(0,60,15)));
    minute_labels = ["XX:00", "XX:15","XX:30","XX:45"]
    ax.set_xticklabels(minute_labels);
    plt.xticks(rotation = 45);
    ax.set_title("Density of Minute in the Hour Stop Occurred")
    ax.set_xlim(0,60);
```

```python
# Make a histogram of time of stops by where in the day they occurred
plot_hist_of_stops_in_day(recent_stops, bins_per_hour=6)
```

```python
density_by_minute_in_hour(recent_stops)
```

```python
## look at non vehicular stops by restricting to "CallType" in [1194, 1194B]
pedestrian_stops = stops[stops['CallType'].isin(['1194','1194B'])]
pedestrian_stops.head()
```

```python
plot_hist_of_stops_in_day(pedestrian_stops)
density_by_minute_in_hour(pedestrian_stops, bins_per_hour=6)
```

```python
# what if we restrict to just "Reas. Sus.", i.e. reasonable suspicion stops 
```

```python
reas_sus_ped_stops = stops[(stops["Reason"]=="Reas. Susp.") & (stops['CallType'].isin(['1194','1194B']))]
reas_sus_ped_stops.head()
```

```python
plot_hist_of_stops_in_day(reas_sus_ped_stops)
density_by_minute_in_hour(reas_sus_ped_stops)
```

```python
recent_reas_ped_stops = reas_sus_ped_stops[reas_sus_ped_stops['year']>=2000]
```

```python
# Make a histogram of time of stops by where in the day they occurred
bins_per_hour = 4
ax = recent_reas_ped_stops[["timeInMinutes"]].hist(figsize=(18,7),bins=24*bins_per_hour, 
                                               density=True, range=(0,24*60))[0][0]
ax.set_xticks(list(range(0,24*60,60)));
hour_labels = ["12:00AM", "1:00AM", "2:00AM", "3:00AM", "4:00AM", "5:00AM", 
                    "6:00AM", "7:00AM", "8:00AM", "9:00AM", "10:00AM", "11:00AM", 
                    "12:00PM", "1:00PM", "2:00PM", "3:00PM", "4:00PM", "5:00PM", 
                    "6:00PM", "7:00PM", "8:00PM", "9:00PM", "10:00PM", "11:00PM"]
ax.set_xticklabels(hour_labels);
plt.xticks(rotation = 45);
ax.set_title("Density of Time of Day Stop Occurred")
ax.set_xlim(0,24*60);
```

```python
# Getting the histogram
bins_per_hour = 6
ax = recent_reas_ped_stops[["minute"]].hist(figsize=(18,7),bins=bins_per_hour, density=True, range=(0,60))[0][0]
ax.set_xticks(list(range(0,60,15)));
minute_labels = ["XX:00", "XX:15","XX:30","XX:45"]
ax.set_xticklabels(minute_labels);
plt.xticks(rotation = 45);
ax.set_title("Density of Minute in the Hour Stop Occurred")
ax.set_xlim(0,60);
```

### Time of Stop Conclusions 

For both Reas. Susp. Pedestrian Stops and overall stops, the time of the stop is fairly uniformly distributed within the hour, which does not support the hypothesis that stops are more likely for officers close to the end of their shifts. Further analysis could involve investigating the actual time of police shifts, improving on the assumption that these are always the same time in the hour.
